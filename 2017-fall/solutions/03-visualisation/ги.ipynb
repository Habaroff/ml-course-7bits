Рассуждение: Посмотрим на графике, как разность между `temp` и `atemp` зависит от `windspeed(ms)`

```python
sns.regplot(x='windspeed(ms)', y='atemp', data=df)
```

```python
df2 = df.copy()
df2['temp_windspeed'] = df2['temp']/(1+df2['windspeed(ms)'])
```

Построим график зависимости нового показателя x=(отношение температуры к скорости ветра) и y=atemp.

Вывод - если жарко и скорость ветра маленькая, то есть x-большое, тем atemp - больше

```python
sns.regplot(x='temp_windspeed', y='atemp', data=df2)
```

```python
df2['temp_minus_atemp'] = df2['temp'] - df2['atemp']
```

Такой получается график, показывающий тенденцию - при усилении ветра разность между `temp` и `atemp` увеличивается (но слабо).

```python
sns.regplot(x='windspeed(ms)', y='temp_minus_atemp' , data=df2)
```

Следующий график показывает, что если жарко и скорость ветра маленькая (т.е. temp_windspeed - больше), то разность между temp и atemp меньше. А вот если температура маленькая и скорость ветра большая (т.е. tamp_windspeed ближе к 0), то разность между temp и atemp -  больше

```python
sns.regplot(x='temp_windspeed', y='temp_minus_atemp' , data=df2)
```
